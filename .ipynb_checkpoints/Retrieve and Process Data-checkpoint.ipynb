### Build Photoshop Dataset using Pandas Dataset

```python
photoshop_dataset = PhotoshopDataset(ORIGINALS_SMALL, PHOTOSHOPS_SMALL)
```

```python
print(len(photoshop_dataset))
```

```python
fig = plt.figure()

num_real = 0
count = 0
for i in range(len(photoshop_dataset)-1, 0, -1):
    sample = photoshop_dataset[i]
    count+=1
    print(i,  sample['tensor'].shape, sample['label'])

    ax = plt.subplot(1, 3, count + 1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(i))
    ax.axis('off')
    image = T.ToPILImage()(sample['tensor'])
    plt.imshow(image)

    if count == 2:
        plt.show()
        break
```
